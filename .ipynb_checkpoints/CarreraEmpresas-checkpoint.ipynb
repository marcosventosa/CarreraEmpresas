```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from pretty_html_table import build_table
import base64
```

```python
fsize = 13
tsize = 18
tdir = 'in'
major = 5.0
minor = 3.0
lwidth = 0.8
lhandle = 2.0
plt.style.use('default')
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family']='serif'
plt.rcParams['font.size'] = fsize
plt.rcParams['legend.fontsize'] = tsize
plt.rcParams['xtick.direction'] = tdir
plt.rcParams['ytick.direction'] = tdir
plt.rcParams['xtick.major.size'] = major
plt.rcParams['xtick.minor.size'] = minor
plt.rcParams['ytick.major.size'] = 5.0
plt.rcParams['ytick.minor.size'] = 3.0
plt.rcParams['axes.linewidth'] = lwidth
plt.rcParams['legend.handlelength'] = lhandle
plt.rcParams["figure.figsize"] = (5.7885,4)
```

```python
params = [['F','10k'],['M','10k'],['F','6k'],['M','6k']]
```

```python
r = requests.get('https://www.carreradelasempresas.com/clasificaciones23/buscador_absoluta.php?sex=M&car=6K')
parsed_html = BeautifulSoup(r.content)
c = parsed_html.body.find('div', attrs={'class':'buscar'})
```

```python
df = pd.DataFrame(columns =["Distancia", "Genero", "Posicion", "Nombre", "Empresa", "Tiempo"] )
for g, k in params:
    r = requests.get('https://www.carreradelasempresas.com/clasificaciones23/buscador_absoluta.php?sex={}&car={}'.format(g,k))
    parsed_html = BeautifulSoup(r.content)
    c = parsed_html.body.find('div', attrs={'class':'buscar'})
    for elm in c.find_all('p'):
        pattern = r'<p><span class="rojo">(\d+)<\/span>  (.*) - <strong>(.*)<\/strong> - ([\d:]+)<\/p>'
        match = re.search(pattern, str(elm))
        df = pd.concat([df,pd.DataFrame(data = [[k]+[g]+list(match.groups())], columns =["Distancia", "Genero", "Posicion", "Nombre", "Empresa","Tiempo"] )])
```

```python
df["Empresa"] = df["Empresa"].str.replace("&amp;","&")
df["Nombre"] = df["Nombre"].str.replace("\t"," ")
```

```python
for g, k in params:
    dfF = df[(df["Distancia"]==k) & (df["Genero"] == g)]
    for emp in set(df["Empresa"]):
        file_name_string = your_string = re.sub(r'[\\/*?:"<>|]',"",emp)
        with open('Empresas/Ranking_{}_{}.html'.format(g+k,file_name_string), 'w', encoding="utf-8") as f:
            table = dfF[dfF["Empresa"] == emp][["Nombre","Tiempo"]]
            table = table.reset_index(drop=True).reset_index().rename(columns={"index":'Pos.'})
            table["Tiempo"] = table["Tiempo"].astype(str).map(lambda x: x[7:])
            table["Pos."]+= 1
            f.write(build_table(table, 'blue_light',font_size='14px'))
```

```python
for g, k in params:
    dfG = df[(df["Genero"] == g) & (df["Distancia"] == k)].groupby(['Empresa'])['Tiempo'].agg(['sum','size'])
    dfG = dfG.rename(columns={'size': 'Personas'})
    dfG["Media"] = dfG['sum'].div(dfG['Personas'])
    dfG = dfG.drop(columns = ["sum"]).reset_index()
    dfG["Media"] = dfG["Media"].astype(str).map(lambda x: x[7:15])
    with open('Rankings/Ranking{}.html'.format(g+k), 'w') as f:
        table = dfG[dfG["Personas"]>=10].sort_values(by = "Media").reset_index(drop = True).reset_index().rename(columns={"index":'Pos.'})
        table["Pos."]+= 1
        f.write(build_table(table, 'blue_light',font_size='15px'))
```

```python
for k in set([k for _, k in params]):
    dfF = df[df["Distancia"]==k]
    datosM = (dfF[dfF["Genero"]=="M" ]['Tiempo'].astype('timedelta64[s]') / 60)
    datosF = (dfF[dfF["Genero"]=="F" ]['Tiempo'].astype('timedelta64[s]') / 60)
    print(k)
    datosM.plot(kind = "hist",bins = 30,density=True,color = 'blue',edgecolor = 'black',alpha=0.5,label='_nolegend_')
    datosF.plot(kind = "hist",bins = 30,density=True,color = 'pink',edgecolor = 'black',alpha=0.5,label='_nolegend_')
    plt.axvline(datosM.mean(), color='blue', linestyle='-', linewidth=1.5,label='Media hombres')
    plt.axvline(datosF.mean(), color='pink', linestyle='-', linewidth=1.5,label='Media mujeres')
    plt.xlabel('Minutos')
    plt.ylabel('Densidad de probabilidad')
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().legend(loc="upper right",fontsize=10)
    plt.yticks([])
    plt.show()
```
